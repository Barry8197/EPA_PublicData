```python
import numpy as np
import pandas as pd
import psycopg2
from tkinter import *
```

```python
hostname = '127.0.0.1'
username = 'catalyst'
password = 'Barryr18'
database = 'eia923'

conn = psycopg2.connect( host=hostname, user=username, password=password, dbname=database )
df = pd.read_sql("SELECT * FROM pg_catalog.pg_tables WHERE schemaname != 'pg_catalog' AND schemaname != 'information_schema';", con=conn)
```

```python
numpy_array = np.array(df)
table = []
for i in numpy_array:
    table.append(i[1])
    
for i in range(len(table)):
    print("%2i. %15s" % (i,table[i]))
```

```python
buttons =[]
Names = []
def printtext():
    global year
    string = year.get()
    for item in table:
        count = 0
        for i in range(len(item)):
            if item[i].isnumeric():
                count +=1
        for i in range(len(item)):
            if item[i:i+4] == string:
                Button_create(item)
            elif count < 3:
                if item[i:i+2] == string[-2:]:
                    Button_create(item)

def close_all():
    root2.destroy()
                    
def Button_create(item):
    button = Button(text = item)
    button.configure(command=lambda button=button: print_table(button))
    button.grid()
    buttons.append([button , item])
    
def reset():
    for i in range(len(buttons)):
        buttons[i][0].destroy()
        
def construct():
    if len(buttons) == 0:
        printtext()
    else :
        reset()
        printtext()
        
def print_table(Name):
    for i in range(len(buttons)):
        if Name == buttons[i][0]:
            Name = buttons[i][1]
            break
    root2 = Tk()
    root2.title(Name)
    root2.geometry("750x500")
    global df
    df = pd.read_sql('SELECT * FROM "%s"' % (Name) , con=conn)
    text = Text(root2)
    text.insert(END, str(df.iloc[:,:]))
    text.grid()
    
buttons =[]   
root = Tk()

root.title('EIA923 Data for years 1970 to present')
root.geometry("500x500")

year = Entry(root)
year.grid(row = 0)
year.focus_set()


r = Button(root, text = 'Reset', command = reset)
r.grid(row = 2)

b = Button(root,text='Enter Data Year',command=construct)
b.grid(row = 1)

root.mainloop()
```
