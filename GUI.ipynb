```python
buttons =[] 
Names = []
checkbox = []

def Export_to_CSV(df,Name):
    writer = pd.ExcelWriter('%s.xlsx' % (Name))
    df.to_excel(writer,'Sheet1')
    writer.save()
    os.system("open -a '/Applications/Microsoft Excel.app' '/Users/barryr/EPA_PublicData/%s.xlsx'" % (Name))

def printtext(Year):
    reset()
    string = Year
    if mvar1.get() == 1 :
        root.title("EIA860 Database for years 1990 to present")
        table = table1
        for item in table:
            if string == '2017ER':
                string = '2017'
            else :
                string = string[-4:]
            check = False 
            for i in range(len(item)):
                if item[i:i+4] == string:
                    Button_create(item)
                    check = True
            if check == False :
                if item[-2:] == string[-2:]:
                    Button_create(item) 
    elif mvar1.get() == 2:
        root.title("EIA923 Database for years 1970 to present")
        string = string[:4]
        table = table2
        for item in table:
            count = 0
            for i in range(len(item)):
                if item[i].isnumeric():
                    count +=1
            for i in range(len(item)):
                if item[i:i+4] == string:
                    Button_create(item)
                elif count < 3:
                    if item[i:i+2] == string[-2:]:
                        Button_create(item)
    else :
        table =[]
                    
def Button_create(item):
    button = Button(text = item)
    button.configure(command=lambda button=button: print_table(button))
    if len(buttons) <12:
        button.grid(row = 4 + len(buttons) , column = 0)
    elif len(buttons) <24:
        button.grid(row =len(buttons)-8,  column = 1)
    elif len(buttons) < 36:
        button.grid(row = len(buttons) - 20, column = 2)
    else :
        button.grid(row = len(buttons) - 32 , column = 3)
    buttons.append([button , item])
    
def reset():
    for i in range(len(buttons)):
        buttons[i][0].destroy()
    del buttons[:]
    if len(checkbox) > 1:
        checkbox[0].destroy()
    for item in os.listdir():
        if item.endswith('.xlsx'):
            os.remove(item)
        
def print_table(Name):
    for i in range(len(buttons)):
        if Name == buttons[i][0]:
            Name = buttons[i][1]
            break
    root2 = Tk()
    root2.title(Name)
    root2.geometry("750x500")
    global df1, df2
    if mvar1.get() == 1:
        df = pd.read_sql('SELECT * FROM "%s"' % (Name) , con=conn1)
    elif mvar1.get() == 2:
        df = pd.read_sql('SELECT * FROM "%s"' % (Name) , con=conn2)
    else :
        df =[]
    excel_button = Button(root2, text = 'Export to csv')
    excel_button.configure(command=lambda excel_button=excel_button: Export_to_CSV(df,Name))
    excel_button.grid()
    text = Text(root2)
    text.insert(END, str(df.iloc[:,:]))
    text.grid()
    
def opmenu():
    if mvar1.get() == 1:
        Year = Years_eia860
    elif mvar1.get() == 2:
        Year = Years_eia923
    else :
        Year = []
    selectedYear = StringVar()
    selectedYear.set(Year[0])
    YearMenu = OptionMenu(root, selectedYear, *Year , command = printtext)
    YearMenu.grid(row = 3 )
    printtext(selectedYear.get())
    checkbox.append(YearMenu)

root = Tk()

root.title('Eia860 & Eia923 Database')
root.geometry('1200x500')

mvar1 = IntVar()
check1 = Radiobutton(root, state = ACTIVE, variable = mvar1, value = 1, text = 'Eia860' , command = opmenu)
check1.grid(row=1)

check2 = Radiobutton(root, state = ACTIVE, variable = mvar1, value = 2, text = 'Eia923' , command = opmenu)
check2.grid(row=2)

root.mainloop()
        
```
